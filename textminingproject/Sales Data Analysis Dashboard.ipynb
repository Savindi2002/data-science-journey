```python
import pandas as pd

data = data = pd.read_csv("sales_data_sample.csv", encoding='latin1')

print(data.head())
```

```python
print(data.info())
```

```python
print(data.describe())
```

```python
total_sales = data["SALES"].sum()
print("Total Sales:", total_sales)
```

```python
sales_product = data.groupby("PRODUCTLINE")["SALES"].sum()
print(sales_product)
```

```python
import matplotlib.pyplot as plt

sales_product.plot(kind='bar')

plt.title("Sales by Product Line")
plt.xlabel("Product Line")
plt.ylabel("Sales")

plt.show()
```
