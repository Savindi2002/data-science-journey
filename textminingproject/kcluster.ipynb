```python
X = [
    [1, 2], [2, 3], [3, 4],
    [8, 7], [9, 8], [10, 9]
]
```

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

X = [
    [1, 2], [2, 3], [3, 4],
    [8, 7], [9, 8], [10, 9]
]

# Create model
kmeans = KMeans(n_clusters=2)

# Fit model
kmeans.fit(X)

# Get labels
labels = kmeans.labels_

print("Cluster Labels:", labels)
```

```python
x_vals = [i[0] for i in X]
y_vals = [i[1] for i in X]

plt.scatter(x_vals, y_vals, c=labels)
plt.title("K-Means Clustering")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```

```python
import pandas as pd

# Load dataset
data = pd.read_csv("data.csv")

print(data.head())
```

```python
X = data[["Income", "Spending"]]
```

```python
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=2)

kmeans.fit(X)

labels = kmeans.labels_

print("Cluster Labels:", labels)
```

```python
data["Cluster"] = labels

print(data.head())
```

```python
import matplotlib.pyplot as plt

plt.scatter(data["Income"], data["Spending"], c=data["Cluster"])

plt.xlabel("Income")
plt.ylabel("Spending")
plt.title("Customer Segmentation")

plt.show()
```

```python
centroids = kmeans.cluster_centers_
print("Centroids:", centroids)

plt.scatter(centroids[:,0], centroids[:,1], s=200, marker='X')

plt.show()
```

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import pandas as pd

data = pd.read_csv("data.csv")
X = data[["Income", "Spending"]]

wcss = []

for k in range(1, 6):
    kmeans = KMeans(n_clusters=k)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)

plt.plot(range(1, 6), wcss)
plt.xlabel("Number of Clusters (K)")
plt.ylabel("WCSS")
plt.title("Elbow Method")

plt.show()
```
