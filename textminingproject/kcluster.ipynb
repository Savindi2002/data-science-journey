```python
X = [
    [1, 2], [2, 3], [3, 4],
    [8, 7], [9, 8], [10, 9]
]
```

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

X = [
    [1, 2], [2, 3], [3, 4],
    [8, 7], [9, 8], [10, 9]
]

# Create model
kmeans = KMeans(n_clusters=2)

# Fit model
kmeans.fit(X)

# Get labels
labels = kmeans.labels_

print("Cluster Labels:", labels)
```
